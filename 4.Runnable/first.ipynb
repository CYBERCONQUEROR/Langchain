```python
import random
class NakliLLM:
    def __init__(self):
        print("NakliLLM initialized")

    def predict(self,prompt):

        response_list=[
            'Delhi is the capital of India',
            'IPL is a cricket tournament',
            'Python is a programming language',
            'AI is the future',

        ]

        return {'response':random.choice(response_list)}
```

```python
llm=NakliLLM()
```
